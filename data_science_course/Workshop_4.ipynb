### Problem 1.2

What is the class distribution? (i.e. how many instances of `malignant` and how many `benign`?)

_This function should return a Series named `target` of length 2 with integer values and index =_ `['malignant', 'benign']`

````python
def answer_two() -> pd.Series:
    """
    Returns a Series with 'malignant' and 'benign' as index and the number of corresponding samples as the value.

    Returns:
        Series: A Series with 'malignant' and 'benign' as index and the number of corresponding samples as the value.
    """
    cancer_df = answer_one()
    malignant = len(cancer_df[cancer_df['target'] == 0])
    benign = len(cancer_df[cancer_df['target'] == 1])
    target = pd.Series([malignant, benign], index = ['malignant', 'benign'])
    return target
````

### Problem 1.4

Using `train_test_split`, split `X` and `y` into training and test sets `(X_train, X_test, y_train, and y_test)`.

__Set the random number generator state to 0 using `random_state=0` to make sure your results match the autograder!__

_This function should return a tuple of length 4: `(X_train, X_test, y_train, y_test)`, where:_

- _`X_train` has shape `(426, 30)`_
- _`X_test` has shape `(143, 30)`_
- _`y_train` has shape `(426,)`_
- _`y_test` has shape `(143,)`_

````python
from sklearn.model_selection import train_test_split

def answer_four() -> tuple:
    """
    Splits the dataset into training and test sets.

    Returns:
        tuple: A tuple of length 4: (X_train, X_test, y_train, y_test).
    """
    X, y = answer_three()
    X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 0)
    return X_train, X_test, y_train, y_test
````

### Problem 1.5

_This function should return a `sklearn.neighbors.classification.KNeighborsClassifier`._

````python
from sklearn.neighbors import KNeighborsClassifier

def answer_five() -> KNeighborsClassifier:
    """
    Trains a KNeighborsClassifier with k = 1 and default parameters.

    Returns:
        KNeighborsClassifier: A KNeighborsClassifier instance.
    """
    X_train, X_test, y_train, y_test = answer_four()
    knn = KNeighborsClassifier(n_neighbors = 1)
    knn.fit(X_train, y_train)
    
    return knn    
````

### Problem 1.9

Using the plotting function below to visualize the different predicition scores between _train_ and _test sets_, as well as malignant and benign cells.

````python
import matplotlib.pyplot as plt

def answer_nine() -> None:
    """
    Plot the accuracies of the training and test sets for malignant and benign cancer cells.
    """
    X_train, X_test, y_train, y_test = answer_four()

    # Training data for plotting
    malignant_train_X = X_train[y_train==0]
    malignant_train_y = y_train[y_train==0]
    benign_train_X = X_train[y_train==1]
    benign_train_y = y_train[y_train==1]

    malignant_test_X = X_test[y_test==0]
    malignant_test_y = y_test[y_test==0]
    benign_test_X = X_test[y_test==1]
    benign_test_y = y_test[y_test==1]

    knn = answer_five()

    scores = [knn.score(malignant_train_X, malignant_train_y), 
              knn.score(benign_train_X, benign_train_y), 
              knn.score(malignant_test_X, malignant_test_y), 
              knn.score(benign_test_X, benign_test_y)]

    # Barplot
    plt.figure()
    bars = plt.bar(np.arange(4), scores, color=['red','red','blue','blue'])
    for bar in bars:
        height = bar.get_height()
        plt.gca().text(bar.get_x() + bar.get_width()/2, height*.90, '{0:.{1}f}'.format(height, 2), 
                     ha='center', color='w', fontsize=11)

    plt.xticks([0,1,2,3], ['Malignant\nTraining', 'Benign\nTraining', 'Malignant\nTest', 'Benign\nTest'], alpha=0.8);
    plt.title('Accuracies for Malignant and Benign Cancer Cells', alpha=0.8)

answer_nine()
````

````python
# Use X_train, X_test, y_train, y_test for all of the following questions
from sklearn.model_selection import train_test_split

df = pd.read_csv('assets/fraud_data.csv')

X = df.iloc[:,:-1]
y = df.iloc[:,-1]

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)
````

### Problem 2.2

Using `X_train`, `X_test`, `y_train`, and `y_test` (as defined above), train a dummy classifier that classifies everything as the majority class of the training data. What is the accuracy of this classifier? What is the recall?

_This function should a return a tuple with two floats, i.e. `(accuracy score, recall score)`._

````python
from sklearn.dummy import DummyClassifier
from sklearn.metrics import recall_score

def answer_twelve() -> tuple:
    """
    Train a dummy classifier that classifies everything as the majority class of the training data. 
    Then find the accuracy and recall of this classifier using the testing data. 

    Returns:
        tuple: A tuple of length 2: (accuracy, recall).
    """
    dummy = DummyClassifier(strategy = 'most_frequent').fit(X_train, y_train)
    y_hat = dummy.predict(X_test)
    
    accuracy = dummy.score(X_test, y_test)
    recall = recall_score(y_test, y_hat)
    
    return (accuracy, float(recall))
````

### Problem 2.3

Using `X_train`, `X_test`, `y_train`, and `y_test` (as defined above), train a _XGBoost_ classifer using the default parameters. What is the accuracy, recall, precision, and F1 Score of this classifier?

_This function should a return a tuple with three floats, i.e. `(accuracy score, recall score, precision score, f1 score)`._

````python
from sklearn.metrics import accuracy_score, recall_score, precision_score, f1_score
from xgboost import XGBClassifier

def answer_thirteen() -> tuple:
    """
    Train a XGBoost classifier on the training data and return the accuracy, recall, precision, and F1 score of the classifier on the test data.
    
    Returns:
        tuple: A tuple of length 4: (accuracy, recall, precision, f1).
    """
    xgb = XGBClassifier().fit(X_train, y_train)
    y_hat = xgb.predict(X_test)
    
    accuracy = accuracy_score(y_test, y_hat)
    recall = recall_score(y_test, y_hat)
    precision = precision_score(y_test, y_hat)
    f1 = f1_score(y_test, y_hat)
    
    return (accuracy, float(recall), float(precision), float(f1))
````

````python
answer_thirteen()
````
